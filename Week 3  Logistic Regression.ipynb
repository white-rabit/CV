```python
import numpy as np
import random
```

```python
def inference(w,b,x):
    z = w*x+b
    pred_y = 1/(1+np.exp(-z))
    return pred_y
```

```python
# logistic
def eval_loss(w,b,x,y):
    
    avg_loss = y * np.log(inference(w,b,x)) + (1-y) * np.log(1-inference(w,b,x))
    avg_loss = -avg_loss.mean()
    return avg_loss
```

```python
def gradient(pred_y,gt_y,x):
    diff = pred_y - gt_y
    dw = diff * x
    db = diff
    return dw, db
```

```python
def cal_step_gradient(batch_x_list, batch_gt_y_list, w,b,lr):
    avg_dw,avg_db = 0,0
    batch_size = len(batch_x_list)
    
    pred_y = inference(w,b,batch_x_list)
    dw,db = gradient(pred_y,batch_gt_y_list,batch_x_list)   
    avg_dw = dw.mean()
    avg_db = db.mean()
    w -= lr * avg_dw
    b -= lr * avg_db
    
    return w,b
```

```python
def train(x_list, gt_y_list, batch_size,lr,max_iter):
    w = 0
    b = 0
    num_samples = len(x_list)
    for i in range (max_iter):
        c = np.arange(num_samples)
        a = np.random.choice(c, size=batch_size, replace=True)
        x = x_list[a]
        y = gt_y_list[a]

        w,b = cal_step_gradient(x,y,w,b,lr)
        print('w:{0}, b:{1}'.format(w,b))
        print('loss is {0}'.format(eval_loss(w,b,x,y)))
```

```python
def gen_sample_data():
    w = random.randint(0,10) +  random.random()
    b = random.randint(0,5) + random.random()
    print(w,b)
    num_samples = 100
    x = np.random.randint(0,10,num_samples)*np.random.random(num_samples)
    y1 = w*x+b+np.random.random(num_samples)*np.random.randint(-1,1,num_samples)
    y = 1/(1+np.exp(-y1))
    return x,y,w,b
```

```python
def run():
    x_list,y_list,w,b = gen_sample_data()
    lr = 0.05
    max_iter = 10000
    train(x_list,y_list,50,lr,max_iter)
    
```

```python
if __name__ == '__main__':
    run()
```
